# Rating examples (raw labels)


```python
from pathlib import Path

import numpy as np
import pandas as pd

from aitraf.data_ops.schema import LabelsSchema

PROJECT_ROOT = Path("..").resolve()
DATA_DIR = PROJECT_ROOT / "data"
LABELS_PATH = DATA_DIR / "labels.jsonl"

pd.set_option("display.max_colwidth", None)

SAMPLES_PER_RATING = 2
RNG_SEED = 1
rng = np.random.default_rng(RNG_SEED)
```

```python
labels_df = pd.read_json(LABELS_PATH, lines=True, dtype=LabelsSchema.types)
ratings = sorted(labels_df["execution_score"].dropna().unique())

rows = []
for rating in ratings:
    rating_rows = labels_df[labels_df["execution_score"] == rating]
    if len(rating_rows) < SAMPLES_PER_RATING:
        raise RuntimeError(f"Not enough rows for rating {rating}.")
    sample_idx = rng.choice(len(rating_rows), size=SAMPLES_PER_RATING, replace=False)
    samples = rating_rows.iloc[sample_idx]
    for _, sample in samples.iterrows():
        rows.append(
            {
                "rating": sample["execution_score"],
                "trick": sample["trick"],
            }
        )

pd.DataFrame(rows)
```
