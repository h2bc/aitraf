```python
from pathlib import Path
import pandas as pd
from ultralytics import YOLO
```

#### Load manifest row

```python
CLIPS_DIR = Path("../data/clips")
MANIFEST_PATH = Path("../data/manifests/train.jsonl")
MANIFESTS_DIR = MANIFEST_PATH.parent
YOLO_WEIGHTS = Path("../models/yolo11n-pose.pt")

manifest_df = pd.read_json(MANIFEST_PATH, lines=True)
```

```python
manifest_df[manifest_df["person"] == "Rytis"]
```

```python
sample = manifest_df[manifest_df["person"] == "Rytis"].iloc[1].to_dict()

clip_path = CLIPS_DIR / sample["video_id"]

clip_path
```

```python
model = YOLO(YOLO_WEIGHTS)
```

```python
results = model.predict(
    source=clip_path.as_posix(),
    device="cuda",
    imgsz=640,
    conf=0.5,
    save=True,
)
```
